```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Andy","Amy","James","Jules","Arthur"])
y = np.array([88,66,90,55,77])

plt.bar(x,y)
plt.title("Students' Grades")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

y = np.array([88,66,90,55,77])
mylabels = ["Andy","Amy","James","Jules","Arthur"]

plt.pie(y, labels = mylabels)
plt.title("Students' Grades")
plt.legend(title = "Students:")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

#1
x= np.array([0,1,2,3])
y= np.array([3,8,1,10])

plt.subplot(2,3,1)
plt.bar(x,y)
plt.grid()
plt.title("Bar Graph One")

#2
plt.subplot(2,3,2)
plt.bar(x,y)
plt.title("Bar Graph Two")
plt.grid()
plt.show()
```
